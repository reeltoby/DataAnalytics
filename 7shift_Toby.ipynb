## Merging both data sets


```python
merged = pd.merge(df, tf, on='company_id', how='outer')
merged.head(10)
```

## Dropping the Unnamed columns

```python
drop_columns = merged.drop(["Unnamed: 0_y", "Unnamed: 0_x"], axis=1)
drop_columns.head(10)
```

```python
missing_val = drop_columns.isnull().sum().sort_values(ascending = False)
missing_val.head(15)
```

```python
####### Finding the % missing values in each column from the combined data set
mis_val = drop_columns.isnull().sum()
mis_val_percent = 100 * drop_columns.isnull().sum() / len(drop_columns)
mis_val_percent
```

```python
#### Visualizing the number of missing values on a table

mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
mis_val_table_ren_columns = mis_val_table.rename(
    columns = {0 : 'Missing Values', 1 : '% of Total of Missing Values'})
mis_val_table_ren_columns
```

### A good chunk of the merged data set is filled with null vales
### This is beacuse most trialist did not indicate their compoany name and never converted to full time customers
### We also will assume that the cancellation_date column with null values signifies that the user still has active subscription.
  

```python
### Selecting all columns with trial_data and sign_up data
sign_up = drop_columns[(drop_columns.signup_date.notnull()) & (drop_columns.trial_date)]
sign_up
```

```python
len(tf)
```

```python
## Getting the conversion rate in %
conversion_rate = 100 * sign_up['signup_date'].count()/ len(tf)
print(f'The conversion rate is {conversion_rate}%')
```
